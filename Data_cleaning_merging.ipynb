# Cimate Data: Rainfall


```python
dataframe_rain = pd.read_csv('rainfall.csv')  
dataframe_rain.head()  
```

```python
dataframe_rain.tail()  
```

```python
dataframe_rain.info() 
```

```python
# change the Datatype
dataframe_rain['average_rain_fall_mm_per_year'] = pd.to_numeric(dataframe_rain['average_rain_fall_mm_per_year'],errors = 'coerce')  
dataframe_rain.info()
```

```python
#Dropping empty rows  
dataframe_rain_main=dataframe_rain.dropna()  
dataframe_rain.describe()
dataframe_rain.head()
```

```python
# SHOW COLUMS IN THE ROW
dataframe_yield_main.columns 
```

```python
dataframe_rain_main.columns
```

```python
# Merging  -  combine the yield dataframe with the rain dataframe according to the year and area columns.
   
dataframe_main = pd.merge(dataframe_yield_main, dataframe_rain_main, on=['Area', 'Year'])  
dataframe_main.head()   
```
